```python
tier = 'CHALLENGER'
result = pd.read_csv(f'../Dataset/perMinuteDataset/result/accuracy/{tier}.csv')
df5 = pd.read_csv(f'../Dataset/perMinuteDataset/5min/{tier}.csv')
df6 = pd.read_csv(f'../Dataset/perMinuteDataset/6min/{tier}.csv')
df7 = pd.read_csv(f'../Dataset/perMinuteDataset/7min/{tier}.csv')
df8 = pd.read_csv(f'../Dataset/perMinuteDataset/8min/{tier}.csv')
df9 = pd.read_csv(f'../Dataset/perMinuteDataset/9min/{tier}.csv')
df10 = pd.read_csv(f'../Dataset/perMinuteDataset/10min/{tier}.csv')
df11 = pd.read_csv(f'../Dataset/perMinuteDataset/11min/{tier}.csv')
df12 = pd.read_csv(f'../Dataset/perMinuteDataset/12min/{tier}.csv')
df13 = pd.read_csv(f'../Dataset/perMinuteDataset/13min/{tier}.csv')
df14 = pd.read_csv(f'../Dataset/perMinuteDataset/14min/{tier}.csv')
df15 = pd.read_csv(f'../Dataset/perMinuteDataset/15min/{tier}.csv')
colList = ['GOLD']
def colValueMerge(data, winLose):
    data = data.rename(columns={f'{winLose}_invadeKill': 'invadeKill', f'{winLose}_invadeDeath': 'invadeDeath', 
                                                  f'{winLose}_controlWARDPlaced': 'controlWARDPlaced',
                                                  f'{winLose}_Kill_top': 'Kill_top',f'{winLose}_Kill_jgl': 'Kill_jgl',f'{winLose}_Kill_mid': 'Kill_mid',f'{winLose}_Kill_ad': 'Kill_ad', f'{winLose}_Kill_sup': 'Kill_sup',
                                                  f'{winLose}_Death_top': 'Death_top',f'{winLose}_Death_jgl': 'Death_jgl',f'{winLose}_Death_mid': 'Death_mid',f'{winLose}_Death_ad': 'Death_ad',f'{winLose}_Death_sup': 'Death_sup',
                                                  f'{winLose}_Asisst_top': 'Assist_top',f'{winLose}_Asisst_jgl': 'Assist_jgl',f'{winLose}_Asisst_mid': 'Assist_mid',f'{winLose}_Asisst_ad': 'Assist_ad',f'{winLose}_Asisst_sup': 'Assist_sup',
                                                  f'{winLose}_LV_top': 'LV_top',f'{winLose}_LV_jgl': 'LV_jgl',f'{winLose}_LV_mid': 'LV_mid',f'{winLose}_LV_ad': 'LV_ad',f'{winLose}_LV_sup': 'LV_sup',
                                                  f'{winLose}_CS_top': 'CS_top',f'{winLose}_CS_jgl': 'CS_jgl',f'{winLose}_CS_mid': 'CS_mid',f'{winLose}_CS_ad': 'CS_ad',f'{winLose}_CS_sup': 'CS_sup',
                                                  f'{winLose}_jglCS_top': 'jglCS_top',f'{winLose}_jglCS_jgl': 'jglCS_jgl',f'{winLose}_jglCS_mid': 'jglCS_mid',f'{winLose}_jglCS_ad': 'jglCS_ad',f'{winLose}_jglCS_sup': 'jglCS_sup',
                                                  f'{winLose}_GOLD_top': 'GOLD_top',f'{winLose}_GOLD_jgl': 'GOLD_jgl',f'{winLose}_GOLD_mid': 'GOLD_mid',f'{winLose}_GOLD_ad': 'GOLD_ad',f'{winLose}_GOLD_sup': 'GOLD_sup',
                                                  f'{winLose}_WARDkill': 'WARDkill',f'{winLose}_Inhibitor': 'Inhibitor',f'{winLose}_TOWERkill': 'TOWERkill',f'{winLose}_WARDplaced': 'WARDplaced'})
    for colName in colList:
        data[colName] = (data[f'{colName}_ad']+data[f'{colName}_top']+data[f'{colName}_jgl']+data[f'{colName}_mid']+data[f'{colName}_sup'])
    return data

winOrLose = 'WIN'
df5 = colValueMerge(df5, winOrLose)
df6 = colValueMerge(df6, winOrLose)
df7 = colValueMerge(df7, winOrLose)
df8 = colValueMerge(df8, winOrLose)
df9 = colValueMerge(df9, winOrLose)
df10 = colValueMerge(df10, winOrLose)
df11 = colValueMerge(df11, winOrLose)
df12 = colValueMerge(df12, winOrLose)
df13 = colValueMerge(df13, winOrLose)
df14 = colValueMerge(df14, winOrLose)
df15 = colValueMerge(df15, winOrLose)
```
