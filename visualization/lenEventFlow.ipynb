```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
```

```python
tier = 'CHALLENGER'
df = pd.read_csv(f'../Dataset/perMinuteDataset/result/event/{tier}.csv')
result = pd.read_csv(f'../Dataset/perMinuteDataset/result/accuracy/{tier}.csv')
```

```python
df_transposed = df.transpose()
df_transposed.columns = df_transposed.iloc[0]

df_transposed = df_transposed[1:]
df_transposed
```

```python
dfMean = df.iloc[:,1:].mean()
result2RandomForest = result[result['Model'] == 'RandomForest']
result2LightGBM = result[result['Model'] == 'LightGBM']
result2CatBoost = result[result['Model'] == 'CatBoost']
result2ExtraTrees = result[result['Model'] == 'ExtraTree']
plt.rcParams['font.family'] ='Malgun Gothic'
plt.figure(figsize=(10, 6))
plt.plot(dfMean.index, dfMean.values*2, label='event', marker='o', linestyle='-', linewidth=5)
plt.plot(dfMean.index, result2RandomForest['accuracy_score'], label='RF_Accuracy', marker='o', linestyle='--')
plt.plot(dfMean.index, result2LightGBM['accuracy_score'], label='LGBM_Accuracy', marker='o', linestyle='--')
plt.plot(dfMean.index, result2CatBoost['accuracy_score'], label='CB_Accuracy', marker='o', linestyle='--')
plt.plot(dfMean.index, result2ExtraTrees['accuracy_score'], label='ET_Accuracy', marker='o', linestyle='--')
plt.title(tier)
plt.xlabel('분')
plt.ylabel('이벤트 발생 개수')
plt.legend()
plt.grid(True)
# 그래프 보여주기
plt.show()
```
